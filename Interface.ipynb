```python
from PIL import Image
import numpy as np
import os
import cv2
from keras.models import load_model
import ipywidgets as widgets
import tkinter as tk
from tkinter import filedialog
from PIL import ImageTk, Image
import matplotlib.image as mpimge
import matplotlib.pyplot as plt
import folium 
from tkinter import *
from tkinter import messagebox
```

```python
file_path=""
text=""
```

```python
%%capture
model1 = load_model('VGG19_best_model.h5')
```

```python
def convert_to_array(img):
    im = cv2.imread(img)
    img = Image.fromarray(im, 'RGB')
    image = img.resize((224, 224))
    return np.array(image)
def get_monument_name(label):
    if label==0:
        return "Statue of liberty, USA"
    elif label==1:
        return "Taj Mahal, India"
    elif label==2:
        return "Eiffel Tower, France"
    elif label==3:
        return "Big Ben, England"
    elif label==4:
        return "Petronas Towers, Malaysia"

def predict_monument(file):
    ar=convert_to_array(file)
    ar=ar/255
    label=1
    a=[]
    a.append(ar)
    a=np.array(a)
    score=model1.predict(a)
    label_index=np.argmax(score)
    acc=np.max(score)
    monument=get_monument_name(label_index)
    if (acc>0.9):
        text=monument+" with an accuracy equals to "+str(acc)
        return text
    else:
        text="not a monument"
        return text
```

```python
import webbrowser
from PIL import Image
window = tk.Tk()
bg="#FBFFC5"
bg1="#F4F9A8"
font="Garamond"
img1=Image.open("test\\blank.jpg")
img1 = img1.resize((300,300), Image.ANTIALIAS)
img1 = ImageTk.PhotoImage(img1)
l1=Label(image=img1)
l1.image = img1
l1.place(x=95,y=135)
def open_file():
    file_path=filedialog.askopenfilename()
    return file_path
def upload():
    global file_path
    file_path=open_file()
    img=Image.open(file_path)
    img = img.resize((300,300), Image.ANTIALIAS)
    img = ImageTk.PhotoImage(img)  
    l=Label(image=img)
    l.image = img
    l.place(x=95,y=135)

def predict():
    global file_path
    global text
    text=predict_monument(file_path)
    if text!='not a monument': 
        messagebox.showinfo("Name of the monument", 'The name of the monument is : ' + text)
    else:
        messagebox.showinfo("Name of the monument", 'not a monument')
    return text      
def location():
    global file_path
    global text
    loc=text.split("with")
    if loc[0] == "Statue of liberty, USA ":
        coord=[40.689247, -74.044502]
        name='Statue of liberty, USA.'
    elif loc[0]=="Taj Mahal, India ":
        coord=[27.173891, 78.042068]
        name='Taj Mahal, India.'
    elif loc[0]=="Eiffel Tower, France ":
        coord=[48.858093, 2.294694]
        name='Eiffel Tower, France.'
    elif loc[0]=="Big Ben, England ":
        coord=[51.510357, -0.116773]
        name='Big Ben, England.'
    elif loc[0]=="Petronas Towers, Malaysia ":
        coord=[3.157764, 101.711861]
        name='Petronas Towers, Malaysia.'
    location = folium.Map(location = coord,zoom_start = 15) 
    folium.Marker(coord,popup=name).add_to(location) 
    location.save("location1.html ")
    webbrowser.open("location1.html ")
def clear():
    window.destroy()
    

window.title("Monuments Recognition")
window.geometry('500x600')
window.configure(background=bg)

upload1=Image.open("icons/upload1.png")
upload1 = upload1.resize((100,100), Image.ANTIALIAS)
upload11 = ImageTk.PhotoImage(upload1)
button = tk.Button(window, text='Upload your image', width=100, command=upload,background=bg1,font=font,image=upload11)
button.image = upload1
button.place(x=250-50,y=10)

monument1=Image.open("icons/monument.png")
monument1 = monument1.resize((100,100), Image.ANTIALIAS)
monument1 = ImageTk.PhotoImage(monument1)
button2 = tk.Button(window, text='Get Name', width=100, command=predict,background=bg1,font=font,image=monument1) 
button2.image = monument1
button2.place(x=150-50,y=455)

location1=Image.open("icons/location.png")
location1 = location1.resize((100,100), Image.ANTIALIAS)
location1 = ImageTk.PhotoImage(location1)
button3 = tk.Button(window, text='Get Location', width=100, command=location,background=bg1,font=font,image=location1)
button3.image = location1
button3.place(x=346-50,y=455)

close1=Image.open("icons/close.png")
close1 = close1.resize((15,15), Image.ANTIALIAS)
close1 = ImageTk.PhotoImage(close1)
button4 = tk.Button(window, text='Close', width=25, command=clear,background=bg,font=font,image=close1) 
button4.image = close1
button4.place(x=550-100,y=10)
window.mainloop()
```
